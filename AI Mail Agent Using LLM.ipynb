```python
customer_email = f"""Subject: Disappointing Performance and Overpriced: NebulaStar Epsilon 2025 Laptop
Dear customer care,
I hope this email finds you well. I am writing to share my recent experience with the NebulaStar Epsilon 2025 laptop from Nebula Electronics. Unfortunately, my experience with this device has been overwhelmingly disappointing, especially considering its exorbitant price tag of $1300.
First and foremost, let's address the elephant in the room: performance. Despite its promising specifications on paper, including the NebulaCore i7_15000H processor and 16GB DDR4 NebulaBoost RAM, the actual performance of this laptop falls far short of expectations. Tasks that should be handled with ease, such as multitasking or running moderately demanding applications, are met with frustratingly sluggish performance. Even simple tasks like web browsing and document editing feel painfully slow, making productivity a constant struggle.
The inadequacy of the NebulaStar Epsilon 2025 becomes even more apparent when considering its price point. At $1300, one would expect a laptop that delivers exceptional performance and value for money. However, this laptop fails to justify its steep price tag in any aspect. Comparable laptops from other manufacturers offer superior performance at a fraction of the cost, making the NebulaStar Epsilon 2025 a poor investment for anyone seeking value and performance.
In my opinion, based on its subpar performance and unjustifiably high price, the NebulaStar Epsilon 2025 should be priced much lower, ideally in the range of $300. At this price point, it would at least be more reasonable for what it offers in terms of performance and features.
In conclusion, I cannot recommend the NebulaStar Epsilon 2025 to anyone in search of a reliable and capable laptop. Its underwhelming performance and inflated price make it a regrettable purchase that falls short of meeting even basic user expectations.
Thank you for taking the time to read my feedback. Should you have any further questions or require additional information, please do not hesitate to reach out.
Sincerely,
john doe"""


def product_category(customer_email):
    delimiter = '####'
    internal_prompt = f"""You will be provided with an email regarding a product which will be delimited by {delimiter}. \
You need to categorize the given email into one of these categories: \
product_inquiry - if the email is regarding any questions about the product, \
product_complaint - if the email is regarding any complaints about the product, \
customer_feedback - if the email is regarding any feedback about the product, \
customer_care - if the email is asking to connect with the customer care service of the company.\n

The response should be a single word only, and that should be the category such as product_inquiry, product_complaint, customer_feedback, or customer_care.
"""
    user_prompt_category = f"""{internal_prompt} {delimiter} {customer_email} {delimiter}"""
    # print(user_prompt_category)
    category_output = model.generate_content(user_prompt_category)
    category_final = category_output.candidates[0].content.parts[0].text
    #print('\n')
    # print("Category:", category_final)
    return category_final

def product_model(customer_email):
    delimiter = '####'
    internal_prompt = f"""You will be provided with an email regarding a product which will be delimited by {delimiter}. \
You need to fetch the product details from the given email. \
From the given email just fetch the model detail and provide back.
"""
    user_prompt_product = f"""{internal_prompt} {delimiter} {customer_email} {delimiter}"""
    product_detail_output = model.generate_content(user_prompt_product)
    product_detail_final = product_detail_output.candidates[0].content.parts[0].text
    #print('\n')
    # print("Product Detail:", product_detail_final)
    return product_detail_final

def fetch_entire_data(product_detail_final, company_product_list):
    delimiter = '####'
    internal_prompt = f"""You will be provided with the model name of the laptop which is {product_detail_final}. \
You need to check for the full value in the given JSON and just pass the JSON of the given model of the laptop. \
You can use this list to fetch the JSON: {company_product_list}.
"""
    regex_ai = f"""{internal_prompt} """
    regex_ai_output = model.generate_content(regex_ai)
    regex_ai_final = regex_ai_output.candidates[0].content.parts[0].text
    #print('\n')
    # print("JSON of the Model:", regex_ai_final)
    return regex_ai_final

def generate_response(category_final, customer_email, regex_ai_final):
    delimiter = '####'
    internal_prompt = f"""The email category which the mail we got from the user belongs to: {category_final}. \
We need to give the response back to the customer email based on: {customer_email}. \
We need to address the mail, write an email to convince the customer that we will be solving their issue or listening to them. \
The data of the product the customer is talking about is this: \
{delimiter}{regex_ai_final}{delimiter}
Our Name is: Jokasda AI
"""
    generate_mail_prompt = f"""{internal_prompt}"""
    generate_mail_output = model.generate_content(generate_mail_prompt)
    # #print('\n')
    print("Generated Response:")
    print(generate_mail_output.candidates[0].content.parts[0].text)

# Example usage:
customer_email = customer_email
company_product_list = company_product_list

# Step 1: Categorize the email
category_final = product_category(customer_email)

# Step 2: Extract product model from email
product_detail_final = product_model(customer_email)

# Step 3: Fetch entire data based on the product model
regex_ai_final = fetch_entire_data(product_detail_final, company_product_list)

# Step 4: Generate response based on category, email, and fetched data
generate_response(category_final, customer_email, regex_ai_final)
```
